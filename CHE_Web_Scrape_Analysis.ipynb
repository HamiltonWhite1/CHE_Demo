```python
import pandas as pd
```

```python
che_csv = pd.read_csv('CHE_scrape.csv')
che_json = pd.read_json('CHE_scrape.json')
cleaned_DF = []
```

```python
for value in che_json['user']:
    normalized = pd.json_normalize(value)
    cleaned_DF.append(normalized)
    
tweets = che_json['full_text'].reset_index(drop=True)
replies = che_json['reply_count'].reset_index(drop=True)
retweets = che_json['retweet_count'].reset_index(drop=True)
favorites = che_json['favorite_count'].reset_index(drop=True)
created = che_json['created_at'].reset_index(drop=True)
url = che_json['url'].reset_index(drop=True)
    

cleaned_DF = pd.concat(cleaned_DF)

# cleaned_DF.merge(tweets, left_on='name', right_on='full_text')
```
